```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Sangley.csv', parse_dates=['Date'], index_col='Date')

# Draw Plot
def plot_df(df, x, y, title="", xlabel='Date', ylabel='Rainfall'): #, dpi=150):
    #plt.figure(figsize=(16,5), dpi=dpi)
    plt.plot(x, y, color='tab:green')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()

plot_df(df, x=df.index, y=df.Rainfall, title='Sangley Climatic Data')  
```
